```python
def stock_factor_treemap_withtab(factor_dataframe,*factor_names):
    '''
    默认是申万一级分类，每个申万一级分类下选取数值最大的20只股票
    '''
    
    tab = Tab()
    for factor_name in factor_names:
        factor = factor_dataframe[['stock_name',factor_name,'industry_name_sw1']]
        factor = factor.rename(columns={'stock_name':'children',factor_name:'value','industry_name_sw1':'name'})
        grouped = factor.groupby(['name']).head(20).groupby(['name'])
        grouped_value = grouped.apply(lambda df:np.sum(df['value']))
        data_for_treemap = []
        for i,grp in enumerate(grouped):
            group_value,group_name = grouped_value[i],grouped_value.index[i]
            children_name,children_value = grp[1]['children'],grp[1]['value']
            children = []
            for j in range(len(children_name)):
                j_data = {'value':children_value.iloc[j],
                         'name':[group_name+'-'+children_name.iloc[j]]}
                children.append(j_data)
            i_data = {'value': group_value,
                     'name':group_name,
                     'children':children}
            data_for_treemap.append(i_data)
        c = (
        TreeMap()
        .add(factor_name, data_for_treemap)
        .set_global_opts(title_opts=opts.TitleOpts(title=factor_dataframe)))
        tab.add(c,factor_name)
    return tab.render_notebook()
```

```python
stock_factor_treemap_withtab(stock_quote_with_indus,'tcap','mcap','amount','turn_rate','chg_value','chg_rate','tclose')
```

```python
stock_factor_treemap_withtab(stock_fin_main_with_indus,'tor','np','roe','np_ratio','gp_ratio','pb','pe','eps','peg')
```
