```python
def get_stock_quote_with_indus_sectional(tdate):
    conn = pymssql.connect(
        server="",
        port="",
        user="",
        password="",
        database="")

    sql = """
    
        if object_id('tempdb..#stock_quote_sw2') is not null drop TABLE #stock_quote_sw2
    select [tdate]
      ,a.[stock_code]
      ,a.[stock_name]
      ,[tcap]
      ,[mcap]
      ,[volume]
      ,[amount]
      ,[turn_rate]
      ,[chg_value]
      ,[chg_rate]
      ,[amplitude]
      ,[high]
      ,[low]
      ,[topen]
      ,[tclose]
      ,[lclose]
      ,[average]
      ,[ltdr]
      ,[r]
      ,[bouns]
      ,[added_bouns]
      ,[tshare]
      ,[mshare] 
  ,b.[standard_code] standard_code_sw2,b.[industry_code] industry_code_sw2,b.[industry_name] industry_name_sw2 into #stock_quote_sw2
from [MONGO_STOCK_DB].[dbo].[ts_stk_quote_daily] A
  inner join [GG_DB].[dbo].[ts_ind_stock] B on a.stock_code = b.stock_code
  where tdate= cast(cast("""+tdate+""" as varchar) as datetime)
  and b.standard_code ='sw2'
  and b.is_last = 1
  and b.status = 1

  
    if object_id('tempdb..#stock_quote_sw1') is not null drop TABLE #stock_quote_sw1
    select [tdate]
      ,a.[stock_code]
      ,a.[stock_name]
      ,[tcap]
      ,[mcap]
      ,[volume]
      ,[amount]
      ,[turn_rate]
      ,[chg_value]
      ,[chg_rate]
      ,[amplitude]
      ,[high]
      ,[low]
      ,[topen]
      ,[tclose]
      ,[lclose]
      ,[average]
      ,[ltdr]
      ,[r]
      ,[bouns]
      ,[added_bouns]
      ,[tshare]
      ,[mshare] 
  ,b.[standard_code] standard_code_sw1,b.[industry_code] industry_code_sw1,b.[industry_name] industry_name_sw1 into #stock_quote_sw1
from [MONGO_STOCK_DB].[dbo].[ts_stk_quote_daily] A
  inner join [GG_DB].[dbo].[ts_ind_stock] B on a.stock_code = b.stock_code
  where tdate= cast(cast("""+tdate+""" as varchar) as datetime)
  and b.standard_code ='sw1'
  and b.is_last = 1
  and b.status = 1


    select a.* ,b.standard_code_sw2,b.industry_code_sw2,b.industry_name_sw2
 from #stock_quote_sw1 a
inner join   #stock_quote_sw2 b
on a.stock_code=b.stock_code
"""
    stock_quote_with_indus = pd.read_sql(sql,conn)
    conn.close()
    return stock_quote_with_indus

stock_quote_with_indus = get_stock_quote_with_indus_sectional('20200930')
```

```python
test = stock_quote_with_indus[['stock_name','tcap','industry_name_sw1']]
```

```python
test = test.rename(columns={'stock_name':'children','tcap':'value','industry_name_sw1':'name'})
```

```python
test
```

```python
grouped = test.groupby(['name']).head(20).groupby(['name'])
```

```python
grouped
```

```python
grouped_value = grouped.apply(lambda df:np.sum(df['value']))
```

```python
grouped_value
```

```python
data_for_treemap = []
```

```python
for i,grp in enumerate(grouped):
    group_value,group_name = grouped_value[i],grouped_value.index[i]
    children_name,children_value = grp[1]['children'],grp[1]['value']
    children = []
    for j in range(len(children_name)):
        j_data = {'value':children_value.iloc[j],
                 'name':[children_name.iloc[j]]}
        children.append(j_data)
    i_data = {'value': group_value,
             'name':group_name,
             'children':children}
    data_for_treemap.append(i_data)
    
```
